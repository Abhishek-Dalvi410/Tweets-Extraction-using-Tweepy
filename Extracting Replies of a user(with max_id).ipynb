```python
import tweepy
import sys
```

```python
consumer_key = '_________'
consumer_secret = '_______________________'
access_token = '_______________________'
access_token_secret = '_______________________'

# OAuth process, using the keys and tokens
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
 
# creation of the actual interface, using authentication
api = tweepy.API(auth,wait_on_rate_limit=True, wait_on_rate_limit_notify=True)
reply=[]
```

```python
def my_awesome_function(id):
    print(id)
    tweets=api.search(q="to:AOC",since_id=id-1,count=200,
                      include_rts=False,max_id=1169249774526185478)
    c=0
    max1=0;
    for i in tweets:
        if (i.in_reply_to_status_id_str==str(id)):
            #print(i.id)
            c=c+1
            max1=i.id
            reply.append(i)
            break;
    if c==0:
        max1=tweets[-1].id
        max1
    counter=0
    while (counter<21):
        tweets123=api.search(q="to:AOC",since_id=id-1,count=200,
                      include_rts=False,max_id=max1-1)
        for inside_tweet in tweets123:
            if(counter==21):
                break
            if(inside_tweet.in_reply_to_status_id_str==str(id)):
                counter=counter+1
                reply.append(inside_tweet)
                print(counter)
        max1=tweets123[-1].id
```
